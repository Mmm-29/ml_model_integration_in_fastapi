```python
import pandas as pd
import numpy as np
df=pd.read_csv("C:/Users/MODAN MOHAN MANNA/Downloads/insurance.csv")
df.head()
```

```python
#feature  engineering
#1. we will combine weight and height to create a new feature bmi
#2. we will make some sge group
#3. we will combine the bmi and smoker to create a new feature lifestyle_risk
#4. we will make some city tier 
```

```python
df["bmi"]=df["weight"]/(df["height"]**2)
df.drop(["height","weight"],inplace=True,axis="columns")
df
```

```python
df["age_group"]=df["age"].apply(
    lambda age:"young" if age < 25 else
    "adult" if age < 45 else 
    "middle_aged" if age < 60 else
    "senior")
df
```

```python
df["lifestyle_risk"]=df.apply(
    lambda row:(
    "high" if row["smoker"] and row["bmi"] > 30  else
    "medium" if row["smoker"] and row["bmi"] > 27 else
    "low"
    ),
    axis="columns" 
)
df
```

```python
tier_1_cities = ["Mumbai", "Delhi", "Bangalore", "Chennai", "Kolkata", "Hyderabad", "Pune"]

tier_2_cities = [
    "Jaipur", "Chandigarh", "Indore", "Lucknow", "Patna", "Ranchi", "Visakhapatnam", "Coimbatore",
    "Bhopal", "Nagpur", "Vadodara", "Surat", "Rajkot", "Jodhpur", "Raipur", "Amritsar", "Varanasi",
    "Agra", "Dehradun", "Mysore", "Jabalpur", "Guwahati", "Thiruvananthapuram", "Ludhiana", "Nashik",
    "Allahabad", "Udaipur", "Aurangabad", "Hubli", "Belgaum", "Salem", "Vijayawada", "Tiruchirappalli",
    "Bhavnagar", "Gwalior", "Dhanbad", "Bareilly", "Aligarh", "Gaya", "Kozhikode", "Warangal",
    "Kolhapur", "Bilaspur", "Jalandhar", "Noida", "Guntur", "Asansol", "Siliguri"
]

# Assign tier to each city
df['city_tier'] = df['city'].apply(
    lambda x:  1 if x in tier_1_cities else
              2 if x in tier_2_cities else
               3
)
df
```

```python
df.drop(["city","smoker","age"],inplace=True,axis="columns")
df
```

```python
x=df.drop(["insurance_premium_category"],axis="columns")
x,x.shape
```

```python
y=df["insurance_premium_category"]
y,y.shape
```

```python
c_col=[]
numeric_col=[]
for col in x.columns:
    if df[col].dtype == "object":
        c_col.append(col)
    else:
        numeric_col.append(col)
```

```python
from sklearn.preprocessing import OrdinalEncoder

enc = OrdinalEncoder(categories=[['Low', 'Medium', 'High']])
y = enc.fit_transform(df[['insurance_premium_category']]).ravel()
y.shape,y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import OneHotEncoder as ohe,StandardScaler as sc
from sklearn.compose import ColumnTransformer as clf
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
```

```python
pipeline=Pipeline(steps=[
    ("preprocessing",clf(
        transformers=[
            ("categorical",ohe(drop="first"),c_col),
            ("numeric",sc(),numeric_col)
        ]
    )),
    ("model",RandomForestClassifier(class_weight='balanced', random_state=42))
])
```

```python
pipeline.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
y_pred=pipeline.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
import pickle
model_path="model.pkl"
with open(model_path,"wb") as file:
    pickle.dump(pipeline,file)
```
